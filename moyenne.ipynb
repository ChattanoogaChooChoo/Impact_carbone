```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming new_data is already created and contains the columns "Country" and "Metric tons of CO2e per capita (2018)"

# Sort the data by CO2 emissions in ascending order
new_data_sorted = new_data.sort_values("Metric tons of CO2e per capita (2018)")

# Add your prediction
X_new = 9
prediction_row = pd.DataFrame({
    "Country": ["Your Prediction"],
    "Metric tons of CO2e per capita (2018)": [X_new]
})
new_data_with_prediction = pd.concat([new_data_sorted, prediction_row]).reset_index(drop=True)

# Create the plot
plt.figure(figsize=(15, 10))
sns.set_style("whitegrid")

# Plot the data points
ax = sns.scatterplot(
    x=range(len(new_data_with_prediction)),
    y="Metric tons of CO2e per capita (2018)",
    data=new_data_with_prediction,
    hue="Metric tons of CO2e per capita (2018)",
    palette="RdYlGn_r",
    legend=False,
    s=100  # Set point size
)

# Connect the points with a line (excluding the prediction point)
plt.plot(range(len(new_data_sorted)), new_data_sorted["Metric tons of CO2e per capita (2018)"], color='gray', alpha=0.5)

# Add the prediction point
prediction_index = len(new_data_with_prediction) - 1
plt.scatter(prediction_index, X_new, color='blue', s=200, marker='*', zorder=5, label='Your Prediction')
plt.annotate('Your Prediction', (prediction_index, X_new), xytext=(5, 5),
             textcoords='offset points', fontsize=10, fontweight='bold')

# Customize the plot
plt.title("CO2 Emissions per Capita (2018) - Ascending Order", fontsize=16)
plt.xlabel("Countries (from least to most polluting)", fontsize=12)
plt.ylabel("Metric tons of CO2e per capita", fontsize=12)

# Improve x-axis labels
plt.xticks(range(len(new_data_with_prediction)), new_data_with_prediction["Country"], rotation=90, ha='right')

# Add value labels
for i, v in enumerate(new_data_with_prediction["Metric tons of CO2e per capita (2018)"]):
    ax.text(i, v, f'{v:.2f}', ha='center', va='bottom', fontsize=8)

# Add a color bar
norm = plt.Normalize(new_data_sorted["Metric tons of CO2e per capita (2018)"].min(),
                     new_data_sorted["Metric tons of CO2e per capita (2018)"].max())
sm = plt.cm.ScalarMappable(cmap="RdYlGn_r", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label("Metric tons of CO2e per capita", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def sample_representative_data(data, n_samples=30):
    """
    Sample approximately n_samples from the data, ensuring representation across the range.
    """
    sorted_data = data.sort_values("Metric tons of CO2e per capita (2018)")

    # Always include the min and max
    result = sorted_data.iloc[[0, -1]]

    if n_samples > 2:
        # Split the remaining range into n_samples-2 intervals
        interval = len(sorted_data) // (n_samples - 1)
        for i in range(1, n_samples - 1):
            idx = i * interval
            result = pd.concat([result, sorted_data.iloc[[idx]]])

    return result.sort_values("Metric tons of CO2e per capita (2018)")

# Assuming new_data is already created and contains the columns "Country" and "Metric tons of CO2e per capita (2018)"

# Sample the data
sampled_data = sample_representative_data(new_data, n_samples=30)

# Add your prediction
X_new = 9
prediction_row = pd.DataFrame({
    "Country": ["Your Prediction"],
    "Metric tons of CO2e per capita (2018)": [X_new]
})
data_with_prediction = pd.concat([sampled_data, prediction_row]).reset_index(drop=True)

# Create the plot
plt.figure(figsize=(15, 10))
sns.set_style("whitegrid")

# Plot the data points
ax = sns.scatterplot(
    x=range(len(data_with_prediction)),
    y="Metric tons of CO2e per capita (2018)",
    data=data_with_prediction,
    hue="Metric tons of CO2e per capita (2018)",
    palette="RdYlGn_r",
    legend=False,
    s=100  # Set point size
)

# Connect the points with a line (excluding the prediction point)
plt.plot(range(len(sampled_data)), sampled_data["Metric tons of CO2e per capita (2018)"], color='gray', alpha=0.5)

# Add the prediction point
prediction_index = len(data_with_prediction) - 1
plt.scatter(prediction_index, X_new, color='blue', s=200, marker='*', zorder=5, label='Your Prediction')
plt.annotate('Your Prediction', (prediction_index, X_new), xytext=(5, 5),
             textcoords='offset points', fontsize=10, fontweight='bold')

# Customize the plot
plt.title("CO2 Emissions per Capita (2018) - Representative Sample", fontsize=16)
plt.xlabel("Countries (sampled from least to most polluting)", fontsize=12)
plt.ylabel("Metric tons of CO2e per capita", fontsize=12)

# Improve x-axis labels
plt.xticks(range(len(data_with_prediction)), data_with_prediction["Country"], rotation=90, ha='right')

# Add value labels
for i, v in enumerate(data_with_prediction["Metric tons of CO2e per capita (2018)"]):
    ax.text(i, v, f'{v:.2f}', ha='center', va='bottom', fontsize=8)

# Add a color bar
norm = plt.Normalize(sampled_data["Metric tons of CO2e per capita (2018)"].min(),
                     sampled_data["Metric tons of CO2e per capita (2018)"].max())
sm = plt.cm.ScalarMappable(cmap="RdYlGn_r", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label("Metric tons of CO2e per capita", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.cm import get_cmap

# List of 30 well-known countries
well_known_countries = [
    "United States", "China", "Japan", "Germany", "United Kingdom",
    "France", "Italy", "Canada", "Russia", "Australia",
    "Brazil", "India", "Mexico", "Spain", "South Korea",
    "Indonesia", "Turkey", "Saudi Arabia", "Switzerland", "Netherlands",
    "Sweden", "Poland", "Belgium", "Norway", "Argentina",
    "Austria", "Thailand", "United Arab Emirates", "South Africa", "Egypt"
]

# Assuming new_data is already created and contains the columns "Country" and "Metric tons of CO2e per capita (2018)"

# Filter the data to include only the well-known countries
filtered_data = new_data[new_data['Country'].isin(well_known_countries)]

# Sort the filtered data by CO2 emissions
sorted_data = filtered_data.sort_values("Metric tons of CO2e per capita (2018)")

# Add your prediction
X_new = 9
prediction_row = pd.DataFrame({
    "Country": ["Vous"],
    "Metric tons of CO2e per capita (2018)": [X_new]
})
data_with_prediction = pd.concat([sorted_data, prediction_row]).sort_values("Metric tons of CO2e per capita (2018)").reset_index(drop=True)

# Create the plot
plt.figure(figsize=(15, 10))
sns.set_style("whitegrid")

# Define a colormap
cmap = get_cmap("RdYlGn_r")

# Normalize the data for color mapping
norm = plt.Normalize(data_with_prediction["Metric tons of CO2e per capita (2018)"].min(),
                     data_with_prediction["Metric tons of CO2e per capita (2018)"].max())

# Create the bar plot with colormap
bars = plt.bar(data_with_prediction["Country"], data_with_prediction["Metric tons of CO2e per capita (2018)"] ,
               color=[cmap(norm(value)) for value in data_with_prediction["Metric tons of CO2e per capita (2018)"]])

# Highlight the prediction bar
prediction_index = data_with_prediction[data_with_prediction["Country"] == "Vous"].index[0]
bars[prediction_index].set_color('black')
bars[prediction_index].set_edgecolor('black')

# Customize the plot
plt.title("Emission moyenne de CO2 d'un individu par pays", fontsize=16)
plt.xlabel("Principaux Pays", fontsize=12)
plt.ylabel("CO2 moyen émis par un individu (en tonne)", fontsize=12)

# Improve x-axis labels
plt.xticks(rotation=90, ha='right')

plt.tight_layout()
plt.show()
```
